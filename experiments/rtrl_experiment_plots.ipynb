```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Humanoid

```python
df_sac_rt = pd.read_pickle('humanoid/sac_humanoid_real_time_stats')
df_sac_no_rt = pd.read_pickle('humanoid/sac_humanoid_no_real_time_stats')
df_rtac = pd.read_pickle('humanoid/rtac_humanoid_real_time_stats')
```

```python
plt.plot(range(len(df_sac_rt['returns'])), df_sac_rt['returns'], label="SAC RT")
plt.plot(range(len(df_rtac['returns'])), df_rtac['returns'], label="RTAC")
plt.plot(range(len(df_sac_no_rt)), df_sac_no_rt['returns'], label="SAC NO RT")
plt.legend()
```

```python
sac_no_rt_cumsum = df_sac_no_rt['returns'].cumsum()
sac_rt_cumsum = df_sac_rt['returns'].cumsum()
rtac_cumsum = df_rtac['returns'].cumsum()

plt.figure(figsize=(8,8))
plt.plot(range(len(df_sac_rt['returns'])), sac_rt_cumsum, label="SAC RTMDP")
plt.plot(range(len(df_rtac['returns'])), rtac_cumsum, label="RTAC RTMDP")
plt.plot(range(len(df_sac_no_rt)), sac_no_rt_cumsum, label="SAC MDP")
plt.title("Humanoid-v2", fontweight="bold", fontsize=20)
plt.xlabel("Steps (x1000)", fontsize=14)
plt.ylabel("Cumulative return", fontsize=14)
plt.legend()
# plt.savefig("humanoid/humanoid_cum_sum.png")
```
